# Histogram - #1

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 이미지 파일명
filename = 'images/레나.png'

# OpenCV를 사용하여 영상 데이터 로딩
image = cv2.imread(filename)

# RGB 영상으로 변환
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 원본 영상에 대한 히스토그램 계산
hist1 = cv2.calcHist([image], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([image], [1], None, [256], [0, 256])
hist3 = cv2.calcHist([image], [2], None, [256], [0, 256])

print(hist1.shape)
print(hist1[:10])
print(hist1[-10:])
```

# Histogram Stretching - #2


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def histogram_stretching(image, ):
  # 원본 히스토그램
  src = cv2.calcHist([image], [0], None, [256], [0, 256])

  # 히스토그램 스트레칭
  image2 = cv2.normalize(image, None, 0, 255, cv2.NORM_MINMAX)
  
  # 수정 히스토그램
  dst = cv2.calcHist([image2], [0], None, [256], [0, 256])

  return image2, src, dst
```

```python
# 이미지 파일명
filename = 'images/Hawkes.png'

# OpenCV를 사용하여 영상 데이터 로딩
image = cv2.imread(filename)

# Grayscale 영상으로 변환
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 원본 이미지에 대한 히스토그램 및 스트레칭
image2, src, dst = histogram_stretching(image)

# 원본 이미지 및 히스토그램 출력
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1), plt.imshow(image, 'gray', vmin=0, vmax=255)
plt.title('image'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2)
plt.plot(src, color='k')
plt.title('histogram'), plt.xlim([0,256])

# 수정 이미지 및 히스토그램 출력
plt.subplot(2, 2, 3), plt.imshow(image2, 'gray', vmin=0, vmax=255)
plt.title('image'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4)
plt.plot(dst, color='k')
plt.title('histogram'), plt.xlim([0,256])
plt.show()
```

# Histogram Sliding - #4


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def histogram_sliding(image, ):
  # 원본 히스토그램
  src = cv2.calcHist([image], [0], None, [256], [0, 256])

  # 히스토그램 슬라이딩
  image2 = cv2.add(image, 50)

  # 수정 히스토그램
  dst = cv2.calcHist([image2], [0], None, [256], [0, 256])

  return image2, src, dst
```

# Histogram Equalization - #5

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def histogram_equalization(image):
  # 원본 히스토그램
  src = cv2.calcHist([image], [0], None, [256], [0, 256])

  # 누적 히스토그램 생성
  src_cdf = src.cumsum()

  # 히스토그램 사이즈로 정규화
  src_cdf = src_cdf * float(max(src)) / max(src_cdf)

  # 히스토그램 균일화
  image2 = cv2.equalizeHist(image)
  
  # 수정 히스토그램
  dst = cv2.calcHist([image2], [0], None, [256], [0, 256])

  # 누적 히스토그램 생성
  dst_cdf = dst.cumsum()

  # 히스토그램 사이즈로 정규화
  dst_cdf = dst_cdf * float(max(dst)) / max(dst_cdf)
  
  return image2, src, src_cdf, dst, dst_cdf
```

```python
filename = 'images/Hawkes.png'

# OpenCV를 사용하여 영상 데이터 로딩
image = cv2.imread(filename)

# 그레이스케일 영상으로 변환
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 원본 이미지에 대한 히스토그램 균일화
image2, src, src_cdf, dst, dst_cdf = histogram_equalization(image)

# 원본 이미지 및 히스토그램 출력
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1), plt.imshow(image, 'gray')
plt.title('image'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 2)
plt.plot(src, color='k')
plt.plot(src_cdf, color='r')
plt.title('histogram'), plt.xlim([0,256])

# 수정 이미지 및 히스토그램 출력
plt.subplot(2, 2, 3), plt.imshow(image2, 'gray')
plt.title('image'), plt.xticks([]), plt.yticks([])
plt.subplot(2, 2, 4)
plt.plot(dst, color='k')
plt.plot(dst_cdf, color='r')
plt.title('histogram'), plt.xlim([0,256])
plt.show()
```

# Contrast Limited Adatptive Histogram Equalization - #6

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def histogram_equalization(image):
  # 원본 히스토그램
  src = cv2.calcHist([image], [0], None, [256], [0, 256])

  # 누적 히스토그램 생성
  src_cdf = src.cumsum()

  # 히스토그램 사이즈로 정규화
  src_cdf = src_cdf * float(max(src)) / max(src_cdf)

  # 히스토그램 균일화
  image2 = cv2.equalizeHist(image)
  
  # 수정 히스토그램
  dst = cv2.calcHist([image2], [0], None, [256], [0, 256])

  # 누적 히스토그램 생성
  dst_cdf = dst.cumsum()

  # 히스토그램 사이즈로 정규화
  dst_cdf = dst_cdf * float(max(dst)) / max(dst_cdf)
  
  return image2, src, src_cdf, dst, dst_cdf
```

```python
def CLAHE(image):
  # 원본 히스토그램
  src = cv2.calcHist([image], [0], None, [256], [0, 256])

  # 누적 히스토그램 생성
  src_cdf = src.cumsum()

  # 히스토그램 사이즈로 정규화
  src_cdf = src_cdf * float(max(src)) / max(src_cdf)

  # contrast limit가 2이고 title의 size는 8X8
  clahe = cv2.createCLAHE(clipLimit=2, tileGridSize=(8,8))
  image2 = clahe.apply(image)
  
  # 수정 히스토그램
  dst = cv2.calcHist([image2], [0], None, [256], [0, 256])

  # 누적 히스토그램 생성
  dst_cdf = dst.cumsum()

  # 히스토그램 사이즈로 정규화
  dst_cdf = dst_cdf * float(max(dst)) / max(dst_cdf)
  
  return image2, src, src_cdf, dst, dst_cdf
```

```python
filename = 'images/석고상.png'

# OpenCV를 사용하여 영상 데이터 로딩
image = cv2.imread(filename)

# 그레이스케일 영상으로 변환
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 원본 이미지에 대한 히스토그램 균일화
image2, src, src_cdf, he, he_cdf = histogram_equalization(image)

# 원본 이미지에 대한 CLAHE
image3, src, src_cdf, clahe, clahe_cdf = CLAHE(image)

# 원본 이미지 및 히스토그램 출력
plt.figure(figsize=(10, 15))
plt.subplot(3, 2, 1), plt.imshow(image, 'gray')
plt.title('image'), plt.xticks([]), plt.yticks([])
plt.subplot(3, 2, 2)
plt.plot(src, color='k')
plt.plot(src_cdf, color='r')
plt.title('histogram'), plt.xlim([0,256])

# 히스토그램 균일화 출력
plt.subplot(3, 2, 3), plt.imshow(image2, 'gray')
plt.title('image'), plt.xticks([]), plt.yticks([])
plt.subplot(3, 2, 4)
plt.plot(he, color='k')
plt.plot(he_cdf, color='r')
plt.title('histogram'), plt.xlim([0,256])

# CLAHE 출력
plt.subplot(3, 2, 5), plt.imshow(image3, 'gray')
plt.title('image'), plt.xticks([]), plt.yticks([])
plt.subplot(3, 2, 6)
plt.plot(clahe, color='k')
plt.plot(clahe_cdf, color='r')
plt.title('histogram'), plt.xlim([0,256])

plt.show()
```
