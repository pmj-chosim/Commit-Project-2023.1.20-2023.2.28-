#beautifulsoup으로 크롤링하기기

```
import requests
from bs4 import BeautifulSoup

# URL of the page to scrape
url = 'https://swedu.knu.ac.kr/05_sub/01_sub.html'

# Send a request to the URL and get the HTML content
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

#print(soup)

#왜 안 됨... 몰라

# Find all the post titles and links on the page
posts = soup.find_all('div', {'class': 'list notice'})

print('-----------\n',posts,'\n-----------\n\n')

# Loop through each post and extract the title and link
for post in posts:
    title = post.find('a').text.strip()
    link = post.find('a')['href']
    print('Title:', title)
    print('Link:', link)
    print('-----------------------------')
```
